# **Sentiment Analysis**

```python
# Model
mdlSent = pipeline('sentiment-analysis')

# Parameters
#prmstatement = 'We are happy to go on vacation this spring break.'
prmstatement = ["I have a feeling that, I will be graduating this year with high grades."]
# Predict
mdlSent(prmstatement)
```

```python
# Model
mdlsnt = pipeline('sentiment-analysis', model = 'nlptown/bert-base-multilingual-uncased-sentiment')

# Parameters
prmstatement = 'Das Wetter ist sehr heiß'

# Predict
mdlsnt(prmstatement)
```

# **Emotion Analysis**

```python
mdlemotion = pipeline("text-classification", model="seara/rubert-tiny2-ru-go-emotions")

emotions = ["Her husband is very stingy..",
            "I have a very trustable friends."]

mdlemotion(emotions)
```

```python

emotions2 = [" I do not know that what should I do to get over with my problems." , "I need some assistance from my insructors and my friends." , "I have a very high enthusiasm towards data science"]

mdlemotion(emotions2)
```

# **Question Answering**

```python
from transformers import pipeline

# Modeli başlatma
mdlQa = pipeline("question-answering")

# Dosyayı açma ve okuma
file_path = "bank-names.txt"
with open(file_path, 'r') as f:
    context = f.read()

print(context)

# Soruyu belirleme
question = "What is the title of it?"

# Tahmin yapma
result = mdlQa(question=question, context=context)

# Sonucu yazdırma
print(result)
```
